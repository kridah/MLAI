```python
import datetime as date
import random
import numpy as np
import matplotlib.pyplot as plt


spent_time = []
result = []


def multiplier(rnd):   
    start = date.datetime.now()
    result.append(rnd * (rnd * rnd))
    end = date.datetime.now()
    spent = end - start
    spent_time.append(str(spent))


for i in range(1, 10000):
    rnd = np.random.randint(1, 1000000)
    multiplier(rnd)
    
    
plt.plot(spent_time)
```
